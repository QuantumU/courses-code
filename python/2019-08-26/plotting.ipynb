```python
# Code von hier:

# Pr√§sentation "Python & Data Science"
# Abschnitt "Daten importieren und exportieren"
# Slide "CSV importieren"

import pandas as pd

iris = pd.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
    header=None,
    names=["sepal_length", "sepal_width", "petal_length",
           "petal_width", "name"])

# erste vier Spalten als NumPy-Array
iris_np = iris.to_numpy()[:, :4].astype("float16")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3])

y1 = x*2
y2 = x**2

plt.plot(x, y1)
plt.plot(x, y2)
```

```python
x = np.arange(0, np.pi*2, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.style.use("ggplot")

plt.plot(x, sin, "g--", label="sin(x)")
plt.plot(x, cos, color=(1, 0.7, 0), marker=".", linestyle="", label="cos(x)")

plt.title("Trigonometric functions")
plt.xlabel("x (radians)")
plt.ylabel("y")
plt.legend()
plt.axis("scaled")
plt.yticks([-1, 0, 1])
plt.xticks(np.linspace(0, 2*np.pi, 5))
```

```python
# Glockenkurve

x = np.linspace(-4, 4, 50)

y = 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x**2)
```

```python
plt.bar(x, y)
plt.plot(x, y, "g-")
```
