## NumPy

```
import numpy as np
```

```
# Python - Listen (mit Verweisen auf andere Integer)
list_a = [1, 2]
list_b = [3, 4]
```

```
# NumPy - Array -
# Daten sind hierin enthalten, ohne auf Python-Integer
# zu verweisen
array_a = np.array(list_a, dtype="int8")
array_b = np.array(list_b, dtype="int8")
```

```
# Python
result = []
for i in range(2):
    result.append(list_a[i] + list_b[i])
print(result)
```
